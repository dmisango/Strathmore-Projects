##Question 3:
Is there a relationship between two columns of your data that you didn't expect? How would you explain it?


```python
#correlations between columns to identify unexpected relationships
correlation_matrix = df1[['Total/Blended Fare', 'Total/Blended Rev', 'Total/Pax']].corr()

# Display the correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for Blended Fare, Blended Revenue, and Pax')
plt.show()
```

The correlation matrix and heatmap highlight a strong positive correlation between 'Blended Revenue' and 'Pax,' suggesting that increased passenger numbers drive higher revenue, as anticipated in transportation data. This relationship is likely driven by factors such as increased ticket sales, higher ancillary revenue, and economies of scale.<br>
However the very low correlation between blended fare and blended revenue is schocking. The expectation was that the higher the fare the higher the revenue. These two have a very low negative correlation suggesting that more often than not a rise or fall in the blended fare will not affect the revenue. The blended fare seems to have a higher relationship with pax than revenue

How does passenger share percentage fluctuate seasonally?

```python
# Calculating the average passenger share percentage for each month
monthly_avg_pax_share = df.filter(like='Pax Share %').mean()
monthly_avg_pax_share = monthly_avg_pax_share*100
monthly_avg_pax_share = monthly_avg_pax_share.sort_values(ascending=False)
# Display the average passenger share percentage for each month
print(monthly_avg_pax_share)
```
